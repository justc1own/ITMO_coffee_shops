```python
# получаем гео пандас с локациями
import requests
import pandas as pd

overpass_url = "https://overpass-api.de/api/interpreter"
query = """
[out:json][timeout:180];
area[name="Санкт-Петербург"]->.searchArea;
(
  node["railway"="station"]["station"="subway"](area.searchArea);
  way["railway"="subway"](area.searchArea);
  relation["route"="subway"](area.searchArea);
);
out body;
"""

response = requests.post(overpass_url, data=query)
data = response.json()

print(data)
```

```python
stations = []
names_station__ = []
for element in data["elements"]:
    if element.get("type") != "node":
        continue

    station = {
        "lat": element.get("lat"),
        "lon": element.get("lon")
    }
    #tags = element.get("tags", {})
    station["name"] = element.get("tags", {}).get("name").replace('-', ' ')
    stations.append(station)
    names_station__.append(station["name"])

geo_df = pd.DataFrame(stations)

geo_df = geo_df.dropna(subset=["lat", "lon"])

print(f"Найдено станций: {len(geo_df)}")
print(geo_df.head())
```

```python
result = {}
print(names_station__)

with open("./parsed_text.txt", 'r') as file:
    text = file.read()

print(text)

for i in range(0, 100):
    print(text[i], end=" ")

for name in names_station__:
    index = text.find(name)
    tmp = ""
    #index += (len(name)+1)
    print("name=", name, " index=", index)

    for i in range(0, len(name)):
        print(text[i+index], end='')
    index += len(name)
    while not (text[index].isalpha()):
        tmp += text[index]
        index += 1

    #print(" ============= val=", val)
    val = tmp.replace(" ", "")
    result.update({name: val})
    print()
print(result)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(result, orient='index', columns=['Value'])
df = df.reset_index().rename(columns={'index': 'name'})
df.columns
```
